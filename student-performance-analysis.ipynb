```python
import pandas as pd
import matplotlib.pyplot as plt

# Create sample student data
data = {
    "Student": ["A", "B", "C", "D", "E", "F", "G", "H"],
    "Study_Hours": [2, 4, 1, 5, 3, 6, 2, 4],
    "Math": [55, 78, 48, 85, 65, 90, 60, 75],
    "Science": [60, 82, 50, 88, 70, 92, 63, 80],
    "English": [70, 85, 65, 90, 75, 95, 72, 88]
}

df = pd.DataFrame(data)
df
```

```python
# Average score per subject
avg_scores = df[["Math", "Science", "English"]].mean()
print("Average Scores:")
print(avg_scores)
```

```python
plt.scatter(df["Study_Hours"], df["Math"])
plt.xlabel("Study Hours")
plt.ylabel("Math Score")
plt.title("Study Hours vs Math Score")
plt.show()
```

```python
correlation = df["Study_Hours"].corr(df["Math"])
print("Correlation between study hours and math score:", correlation)
```
